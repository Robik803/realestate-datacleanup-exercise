#### Exercise 07. Which is the mean of prices in the population (level5 column) of "Arroyomolinos (Madrid)"? (★★☆)

#### Exercise 09. Are the average prices of "Valdemorillo" and "Galapagar" the same? (★★☆)

Print both average prices and then write a conclusion about them.

```python
prices = ds[["price","level5"]].groupby("level5").mean()
valdemorillo = prices.loc[prices.index == "Valdemorillo"].iloc[0,0]
galapagar = prices.loc[prices.index == "Galapagar"].iloc[0,0]
print(f"The average price of Valdemorillo is {round(valdemorillo,2)} USD")
print(f"The average price of Galapagar is {round(galapagar,2)} USD")
print(f"The average prices for these two populations are almost the same ({round(abs(valdemorillo-galapagar),2)} USD difference)")
```

#### Exercise 10. Are the average prices per square meter (price/m2) of "Valdemorillo" and "Galapagar" the same? (★★☆)

Print both average prices and then write a conclusion about it.

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
level5Prices = ds[["price","surface","level5"]]
level5Prices["pps"] = level5Prices["price"]/level5Prices["surface"]
prices = level5Prices.groupby("level5").mean()
valdemorillo = prices.loc[prices.index == "Valdemorillo","pps"].iloc[0]
galapagar = prices.loc[prices.index == "Galapagar","pps"].iloc[0]
print(f"The average price per square meter of Valdemorillo is {round(valdemorillo,2)} USD")
print(f"The average price per square meter of Galapagar is {round(galapagar,2)} USD")
print(f"The average price per square meter for Valdemorillo is lower than Galapagar's price")
```

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in Exercise 14 (★★☆)

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
level5Prices = southBelt[["price","surface","level5"]]
level5Prices["pps"] = level5Prices["price"]/level5Prices["surface"]
prices = level5Prices.groupby("level5").mean()
getafe = prices.loc[prices.index == "Getafe","pps"].iloc[0]
alcorcon = prices.loc[prices.index == "Alcorcón","pps"].iloc[0]
print(f"The average price per square meter of Getafe is {round(getafe,2)} USD")
print(f"The average price per square meter of Alcorcón is {round(alcorcon,2)} USD")
print("The average price per square meter for Getafe is lower than Alcorcón's price")
```
